### ¿Qué es REST web service y cómo se utilizan?

REST (Representational State Transfer) es un estilo de arquitectura para diseñar servicios web. Se basa en un conjunto de principios que definen cómo se deben estructurar y acceder a los recursos en la web. Los servicios web que siguen estos principios se conocen como RESTful web services. 

Los REST web services funcionan a través de HTTP y utilizan sus métodos (GET, POST, PUT, DELETE, etc.) para realizar operaciones CRUD (Crear, Leer, Actualizar, Eliminar) en los recursos. Cada recurso en un REST web service se identifica de manera única a través de URIs (Uniform Resource Identifiers). Por ejemplo, en un servicio web RESTful que gestiona un sistema de libros en una biblioteca, podría haber una URI como `http://ejemplo.com/libros/1` para acceder al libro con ID 1.

Los REST web services son ampliamente utilizados debido a su simplicidad, escalabilidad, y compatibilidad con la web, lo que los hace ideales para aplicaciones web, servicios en la nube, y APIs que son consumidas por diversas plataformas y dispositivos.

### ¿Qué es una RESTful API?

Una RESTful API es una interfaz de programación de aplicaciones que sigue los principios de REST. Permite la comunicación entre diferentes sistemas de software en la web. Las RESTful APIs utilizan métodos HTTP para realizar las operaciones CRUD sobre los recursos representados, usualmente en formatos como JSON o XML.

Las RESTful APIs son diseñadas para ser fáciles de entender y usar, y están orientadas a los recursos, lo que significa que cada API está diseñada en torno a los recursos del sistema y cómo se accede y modifica estos recursos. Esto hace que las RESTful APIs sean muy flexibles y escalables, y son una elección popular para construir servicios web públicos y privados.

### ¿Qué es JSON?

JSON (JavaScript Object Notation) es un formato ligero de intercambio de datos, fácil de leer y escribir para los humanos, y fácil de analizar y generar para las máquinas. Está basado en un subconjunto del lenguaje de programación JavaScript, aunque es independiente del lenguaje y se puede utilizar en muchos entornos de programación diferentes.

JSON está compuesto por dos estructuras:
1. Una colección de pares clave/valor (a menudo llamada un objeto en otros lenguajes de programación).
2. Una lista ordenada de valores (a menudo llamada un array, vector, lista o secuencia en otros lenguajes de programación).

Debido a su simplicidad y facilidad de uso, JSON se ha convertido en el formato estándar para el intercambio de datos en la web, especialmente en las RESTful APIs, donde los recursos son frecuentemente representados y transmitidos en formato JSON.

Para encontrar el total de naves y cuántas de ellas están activas, primero analizaremos el total de naves únicas en el DataFrame y luego filtraremos por su estado de actividad para entender cuántas de ellas están en servicio activo.

#### Total de Naves

Para determinar el total de naves en el DataFrame, contamos la cantidad de nombres únicos de naves, ya que cada nave tiene un nombre distinto. Este enfoque asegura que no contemos la misma nave más de una vez.

```python
#Cantidad total de naves
#Conteo de valores
conteo_naves = df['name'].value_counts()
#Definir la cantdad total de valores distintos
cantidad_naves = len(conteo_naves)
cantidad_naves
```

El resultado muestra que hay un total de 29 naves en el DataFrame. Este número representa la diversidad de la flota disponible, incluyendo diferentes tipos y funciones de naves.

#### Naves Activas

Para encontrar el número de naves activas, agrupamos el DataFrame por la columna 'active' y contamos el número de entradas en cada grupo. Este método nos permite ver cuántas naves están activas (True) y cuántas no están activas (False).

```python
#Cantidad de nave activas
conteo_activas = df.groupby('active').size()
conteo_activas
```

La salida indica que hay 15 naves activas y 14 naves no activas en el DataFrame. Esto refleja que una buena parte de la flota está actualmente en servicio y contribuyendo a las operaciones y misiones, mientras que una cantidad similar está inactiva, lo que podría deberse a mantenimiento, retiro o reserva para futuras misiones.
